```python
import glob
import os
import pandas as pd
import re
from unicodedata import normalize
from tqdm.auto import tqdm
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext
```

```python
def clean(path):
    text={"title":"","context":"","url":""}
    page_view_line = 0
    with open(path, 'r', encoding='utf-8-sig') as f:
        for n, line in enumerate(f):
            line = line.strip()
            if n == 0: # title line
                text["title"] = line.strip()
            else:
                if line:
                    if re.match(r'^[\d,]+$', line):
                        page_view_line = n
                        continue
                    if line == 'พิมพ์' or page_view_line and page_view_line < n: # skip 'print'
                        continue
                    if re.match(r'^ที่มา : http', line):
                        text["url"] = line.strip().split(" ")[-1]
                    else:
                        text["context"] +=line.strip().replace("\xa0","")+"\n"
    text["context"]=cleanhtml(normalize('NFKD', text["context"])).strip()
    text["raw"]=text["title"]+"\n"+text["context"]
    return text
```

```python
clean("./thaigov2-9/ข่าวทันสถานการณ์โควิด-19_1.txt")
```

```python
list_text = list(glob.glob(os.path.join("*","*.txt")))+list(glob.glob(os.path.join("*","*","*.txt")))+list(glob.glob(os.path.join("*","*","*","*.txt")))
```

```python
list_text=list(set(list_text))
```

```python
len(list_text)
```

```python
list_data=[]
for i in tqdm(list_text):
    list_data.append(clean(i))
```

```python
list_data[-1]
```

```python
title=[]
context=[]
url=[]
raw=[]
for d in tqdm(list_data):
    title.append(d["title"])
    context.append(d["context"])
    url.append(d["url"])
    raw.append(d["raw"])
```

```python
df = pd.DataFrame.from_dict(
    {
        "title": title,
        "context": context,
        "raw":raw,
        "url": url,
    }
)
```
